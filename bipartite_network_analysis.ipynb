```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
## Matplotlib defaults ##
colors = {"avg_degree": "#aa0f35", 
         "max_degree": "#fdc799",
         "nodes": "#abdbfc",
         "lcc": "#a0a75c",
         "edges": "#603734",
         "cc": "#009ecb",
         "2010": "#430057",
         "2015": "#f89b00"}

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

plt.rcParams.update({'font.size': 22})
```

## Basic network stats

Plotting of basic stats - degree, clustering coefficient, LCC, etc...

```python
#Incomplete data
data = pickle.load(file=open("Data/NetworkStats/updated_networkStats.p", "rb"))
df = pd.DataFrame(data)
print(df.head())
print(len(df))
```

```python
#Full data (currently through 1998...)
df = pd.read_csv("Data/NetworkStats/stats_1980_1998.csv")
```

```python
#Avg Degree
plt.figure(figsize=(6,6))

#These statistics are only through the 9th month of 1986 (1986.75) 
plt.plot(np.arange(1980,1999,1/12), df["Avg Degree"], linewidth=3, color=colors["avg_degree"], label="Average Degree")

#plt.ylim([60,65])
plt.ylabel("Degree")
plt.xlabel("Year")
plt.legend(prop={"size":14})
```

```python
#Max Degree - boring lol
plt.figure(figsize=(6,6))

#These statistics are only through the 9th month of 1986 (1986.75) 
plt.plot(np.arange(1980,1999,1/12), df["Max Degree"], linewidth=3, color=colors["max_degree"], label="Max Degree")

#plt.ylim([60,65])
plt.ylabel("Degree")
plt.xlabel("Year")
plt.legend(prop={"size":14})
```

```python
#Network Size & LCC size

plt.figure(figsize=(8,8))

#These statistics are only through the 9th month of 1986 (1986.75) 
plt.plot(np.arange(1980,1999,1/12), df["Nodes"], linewidth=3, color=colors["nodes"], label="Total Nodes")
plt.plot(np.arange(1980,1999,1/12), df["LCC Size"], linewidth=3, color=colors["lcc"], label="Largest Connected Cluster Size")

#plt.ylim([60,65])
plt.ylabel("Nodes")
plt.xlabel("Year")
plt.legend(prop={"size":14})
```

```python
plt.figure(figsize=(6,6))

#These statistics are only through the 9th month of 1986 (1986.75) 
plt.plot(np.arange(1980,1999,1/12), df["Edges"], linewidth=3, color=colors["edges"], label="Total Edges")

#plt.ylim([60,65])
plt.ylabel("Edges")
plt.xlabel("Year")
plt.legend(prop={"size":14})
```

```python
#Average clustering coefficient - all zero, boring again lol
plt.figure(figsize=(6,6))

#These statistics are only through the 9th month of 1986 (1986.75) 
plt.plot(np.arange(1980,1999,1/12), df["Clustering coefficient"], linewidth=3, color=colors["cc"], label="Avg Clustering Coefficient")

#plt.ylim([60,65])
plt.ylabel("Clustering Coefficient")
plt.xlabel("Year")
plt.legend(prop={"size":14})
```
